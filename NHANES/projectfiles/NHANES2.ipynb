```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
years = [("1999-2000", ""), ("2001-2002", "_B"), ("2003-2004", "_C"), ("2005-2006", "_D"), ("2007-2008", "_E"),
        ("2009-2010", "_F"), ("2011-2012", "_G"), ("2013-2014", "_H"), ("2015-2016", "_I"), ("2017-2018", "_J")]

# one is mutible and the other you cant. tuples and a list.

conditions = ["BPX"]

def get_data(url):
    all_url = []
    for condition in conditions:
        for year in years:
            get_url = f"{url}/{year[0]}/{condition}{year[1]}.XPT"
            all_url.append(get_url)
    # add the 2017-2020 data url manually
    all_url.append("https://wwwn.cdc.gov/Nchs/Nhanes/2017-2018/P_BPXO.XPT")
    return all_url

base_url = "https://wwwn.cdc.gov/Nchs/Nhanes"
all_extracted_url = get_data(base_url)
```

```python
# not everything in same
# coconcate only combines the same column names.
# change the name of the last column and then concatenate them.

col_of_interest = ["BPXSY1", "BPXSY2", "BPXSY3",
                    "BPXDI1", "BPXDI2", "BPXDI3"]

col_of_interest_mapper = {"BPXOSY1": "BPXSY1", "BPXOSY2":"BPXSY2", "BPXOSY3":"BPXSY3",
                         "BPXODI1":"BPXDI1", "BPXODI2":"BPXDI2", "BPXODI3": "BPXDI3"}

for i, data_url in enumerate(all_extracted_url):
    if i == 0:
        df = pd.read_sas(data_url)[col_of_interest]
    elif i == 10: # last url because of osciculate reading
        data = pd.read_sas(data_url).rename(columns = col_of_interest_mapper)
        data = data[col_of_interest]
        df = pd.concat([df,data])
    else:
        data = pd.read_sas(data_url)[col_of_interest]
        df = pd.concat([df,data])
```

```python
col_of_interest = ["BPXSY1", "BPXSY2", "BPXSY3",
                    "BPXDI1", "BPXDI2", "BPXDI3"]

col_of_interest_mapper = {"BPXOSY1": "BPXSY1", "BPXOSY2":"BPXSY2", "BPXOSY3":"BPXSY3",
                         "BPXODI1":"BPXDI1", "BPXODI2":"BPXDI2", "BPXODI3": "BPXDI3"}

for i, data_url in enumerate(all_extracted_url):
    if i == 0:
        df = pd.read_sas(data_url)[col_of_interest]
        df["Year"] = years[i][0] # i is the years and []
    elif i == 10: # last url because of osciculate reading
        data = pd.read_sas(data_url).rename(columns = col_of_interest_mapper)
        data = data[col_of_interest]
        data["Year"] = "2017-2020" #manual change
        df = pd.concat([df,data])
    else:
        data = pd.read_sas(data_url)[col_of_interest]
        data["Year"] = years[i][0]  # data not df
        df = pd.concat([df,data])
```
